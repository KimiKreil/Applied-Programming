```python
import numpy as np
import matplotlib.pyplot as plt

# Load the file
data = np.genfromtxt('xy.dat', delimiter=',')

# Define true solution
x_range = (0,1)
true_x = np.linspace(*x_range, 100)
true_y = np.exp(-true_x)

# Plot it
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(data[:,0], data[:,1], label='Implicit Euler Solution')
ax.plot(true_x, true_y, label='True Solution')
plt.legend()
plt.show()
```
